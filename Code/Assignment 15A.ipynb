# Label encoding

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

company_d["ShelveLoc"] = label_encoder.fit_transform(company_d["ShelveLoc"])
company_d["Urban"] = label_encoder.fit_transform(company_d["Urban"])
company_d["US"] = label_encoder.fit_transform(company_d["US"])
company_d
```

```python
company_d.loc[company_d["Sales"] <= 7.49,"high"]="0"
company_d.loc[company_d["Sales"] >=7.49,"high"]="1"
```

```python
## Train and Split the data 
x = company_d.drop(['high', 'Sales'],axis=1)
y = company_d['high']
```

```python
from sklearn.ensemble import RandomForestClassifier
## Random forest Classification
num_trees = 100
max_features = 4
kfold = KFold(n_splits=10,  shuffle=True)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features)
results_rfc = cross_val_score(model, x, y, cv=kfold)
print(results_rfc.mean())
```

 # Various Ensemble Techniques

```python
# Bagging 
from sklearn.ensemble import BaggingClassifier

seed = 7
kfold = KFold(n_splits=10, random_state=seed, shuffle=True)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
results_bag = cross_val_score(model, x, y, cv=kfold)
print(results_bag.mean())
```

```python
# Boosting 
from sklearn.ensemble import AdaBoostClassifier

num_trees = 100
seed=7
kfold = KFold(n_splits=10, random_state=seed, shuffle=True)
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results_boost = cross_val_score(model, x, y, cv=kfold)
print(results_boost.mean())
```

```python
# Stacking 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
# Creating the sub models
estimators = []
model1 = LogisticRegression(max_iter=500)
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))

# Creating the ensemble model
ensemble = VotingClassifier(estimators)
results_stack = cross_val_score(ensemble, x, y, cv=kfold)
print(results_stack.mean())
```

***Boosting Technique gave the Highest Accracy i.e 0.835***
